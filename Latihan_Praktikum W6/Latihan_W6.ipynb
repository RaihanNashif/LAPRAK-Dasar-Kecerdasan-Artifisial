```python
import networkx as nx
import matplotlib.pyplot as plt

# Buat graf tak berarah dengan bobot
G = nx.Graph()

# Menambahkan edges (termasuk edge Iasi-Neamt)
edges = [
    ("Oradea", "Zerind", 71),
    ("Zerind", "Arad", 75),
    ("Arad", "Timisoara", 118),
    ("Arad", "Sibiu", 140),
    ("Oradea", "Sibiu", 151),
    ("Timisoara", "Lugoj", 111),
    ("Lugoj", "Mehadia", 70),
    ("Mehadia", "Drobeta", 75),
    ("Drobeta", "Craiova", 120),
    ("Craiova", "Rimnicu Vilcea", 146),
    ("Craiova", "Pitesti", 138),
    ("Rimnicu Vilcea", "Sibiu", 80),
    ("Rimnicu Vilcea", "Pitesti", 97),
    ("Sibiu", "Fagaras", 99),
    ("Fagaras", "Bucharest", 211),
    ("Pitesti", "Bucharest", 101),
    ("Bucharest", "Giurgiu", 90),
    ("Bucharest", "Urziceni", 85),
    ("Urziceni", "Hirsova", 98),
    ("Hirsova", "Eforie", 86),
    ("Urziceni", "Vaslui", 142),
    ("Vaslui", "Iasi", 92),
    ("Iasi", "Neamt", 87),
]

G.add_weighted_edges_from(edges)

# Menentukan posisi node secara manual
pos = {
    "Oradea": (1.5, 9.2),
    "Zerind": (1.3, 7.5),
    "Arad": (1.0, 6.0),
    "Timisoara": (1.0, 3.8),
    "Lugoj": (2.3, 3.5),
    "Mehadia": (2.7, 2.2),
    "Drobeta": (3.3, 1.0),
    "Craiova": (4.8, 2.0),
    "Rimnicu Vilcea": (4.5, 3.5),
    "Pitesti": (5.6, 3.1),
    "Sibiu": (3.2, 6.5),
    "Fagaras": (5.1, 6.5),
    "Bucharest": (6.7, 2.6),
    "Giurgiu": (6.8, 1.2),
    "Urziceni": (7.6, 3.5),
    "Hirsova": (8.8, 3.6),
    "Eforie": (9.5, 2.8),
    "Vaslui": (8.5, 5.2),
    "Iasi": (9.3, 6.4),
    "Neamt": (10.0, 7.5),
}



# Gambar graf 
plt.figure(figsize=(16, 10))
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', 
        node_size=2500, font_size=9, node_shape='s')
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("Graf Kota-Kota Besar di Eropa (Mirip Gambar Soal)", fontsize=14)
plt.axis("off")
plt.show()

# DFS dari Arad
dfs_result = list(nx.dfs_tree(G, source="Arad"))
print("Hasil DFS dari Arad:")
print(dfs_result)
```

Pada algoritma Depth First Search (DFS), informasi weight (bobot) pada graf tidak digunakan dalam proses pencarian. DFS hanya memperhatikan struktur graf dan hubungan antar-node tanpa mempertimbangkan nilai bobot pada setiap edge.

Pada fungsi "nx.dfs_tree(G, source="Arad")" melakukan traversal DFS dari simpul "Arad". DFS bekerja dengan eksplorasi sedalam mungkin sebelum kembali untuk mencari jalur lain. Proses ini dilakukan tanpa mempertimbangkan bobot dari edge yang menghubungkan node-node.

Jadi, hasil DFS yang dihasilkan dari kode ini, urutan kunjungan tidak dipengaruhi oleh bobot pada graf. Hanya struktur konektivitas antar-node yang menentukan jalannya eksplorasi.


```python
import networkx as nx
import matplotlib.pyplot as plt

# Buat graf
G = nx.Graph()

# Tambahkan edge 
edges = [
    ("Jakarta", "Cirebon", 327),
    ("Jakarta", "Bandung", 270),
    ("Bandung", "Cirebon", 120),
    ("Cirebon", "Semarang", 305),
    ("Bandung", "Yogyakarta", 373),
    ("Cirebon", "Yogyakarta", 210),
    ("Semarang", "Yogyakarta", 109),
    ("Yogyakarta", "Surakarta", 60),
    ("Surakarta", "Semarang", 97),
    ("Surakarta", "Malang", 370),
    ("Semarang", "Surabaya", 369),
    ("Surabaya", "Malang", 94)
]

# Tambahkan ke graf
G.add_weighted_edges_from(edges)

# Posisi node 
pos = {
    "Jakarta": (0, 4),
    "Bandung": (2, 1),
    "Cirebon": (5.5, 3.5),
    "Semarang": (13, 3),
    "Yogyakarta": (12.5, 0),
    "Surakarta": (16.5, 1),
    "Surabaya": (25, 2.5),
    "Malang": (24, 0)
}

# Gambar graf
plt.figure(figsize=(12, 6))
nx.draw(
    G, pos, with_labels=True, node_color='lightblue',
    node_size=2500, font_size=9, edge_color='gray' 
)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("Graf Kota-Kota Besar di Indonesia (Pulau Jawa)", fontsize=14)
plt.show()

# Lakukan DFS dari Bandung
dfs_order = list(nx.dfs_preorder_nodes(G, source='Bandung'))
print("Hasil DFS dari Bandung:")
print(dfs_order)
```

Pada algoritma Depth First Search (DFS), informasi weight (bobot) pada graf tidak digunakan dalam proses pencarian. DFS hanya fokus pada struktur graf dan bagaimana node-node saling terhubung.

Dalam kode diatas, di bagian fungsi "dfs_order = list(nx.dfs_preorder_nodes(G, source='Bandung'))", melakukan traversal DFS dari simpul "Bandung" dan mengunjungi node sedalam mungkin sebelum kembali untuk eksplorasi lebih lanjut. Bobot dari jalur antar kota tidak berpengaruh dalam urutan hasil pencarian.

Jadi kesimpulan dari algoritma DFS (Depth-First Search) hanya fokus pada struktur graf yang dimana ia akan memilih satu jalur untuk dieksplor sedalam mungkin, tanpa melihat panjang atau bobot edge tersebut. Selama ada simpul yang belum dikunjungi, DFS akan terus menyelam ke simpul itu. Jika sudah buntu (tidak ada simpul baru), baru DFS akan mundur ke simpul sebelumnya.
